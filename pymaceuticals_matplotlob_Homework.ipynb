```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinical_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
study_data = pd.merge(clinical_data, mouse_data, how="left", on=["Mouse ID"])

# Display the data table for preview
study_data.head()

```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
#store mean data - grouped by drug name and time point
# Convert to DataFrame

# Preview DataFrame
tumor_sizes = study_data.groupby(["Drug", "Timepoint"]).agg({"Tumor Volume (mm3)": "mean"}).reset_index()
tumor_sizes.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
# Preview DataFrame
tumor_sizes_se = study_data.groupby(["Timepoint", "Drug"]).agg({"Tumor Volume (mm3)": "sem"}).reset_index()\
    .pivot(values='Tumor Volume (mm3)', index='Timepoint', columns='Drug')
```

```python
tumor_sizes_se.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_sizes2 = study_data.groupby(["Timepoint", "Drug"]).agg({"Tumor Volume (mm3)": "mean"}).reset_index()\
    .pivot(values='Tumor Volume (mm3)', index='Timepoint', columns='Drug')

# Preview that Reformatting worked
tumor_sizes2.head()
```

```python
# Generate the Plot (with Error Bars)
x_axis = tumor_sizes2.index.values

plt.figure(figsize=(13,6))
plt.errorbar(x_axis, tumor_sizes2["Capomulin"], linestyle=':', linewidth=1, yerr=tumor_sizes_se["Capomulin"],\
         marker="o", color="red", label="Capomulin")
plt.errorbar(x_axis, tumor_sizes2["Infubinol"], linestyle=':', linewidth=1, yerr=tumor_sizes_se["Infubinol"],\
         marker="^", color="blue", label="Infubinol")
plt.errorbar(x_axis, tumor_sizes2["Ketapril"], linestyle=':', linewidth=1, yerr=tumor_sizes_se["Ketapril"], \
         marker="s", color="green", label="Ketapril")
plt.errorbar(x_axis, tumor_sizes2["Placebo"], linestyle=':', linewidth=1, yerr=tumor_sizes_se["Placebo"], \
         marker="D", color="black", label="Placebo")

plt.grid()
plt.legend(loc="best")
plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.tight_layout()

# Save the Figure
plt.savefig("Images/TumorResponse.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint  
meta_sites = study_data.groupby(["Drug", "Timepoint"]).agg({"Metastatic Sites": "mean"})

# Preview DataFrame
meta_sites.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
meta_sites_se = study_data.groupby(["Timepoint", "Drug"]).agg({"Metastatic Sites": "sem"}).reset_index()\
    .pivot(values='Metastatic Sites', index='Timepoint', columns='Drug')

# Preview new DataFrame
meta_sites_se.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
# The values shown in the instructions are incorrect for Metastatic Sites. 
# Those values correspond to Tumor Response Treatment
meta_sites2 = study_data.groupby(["Timepoint", "Drug"]).agg({"Metastatic Sites": "mean"}).reset_index()\
    .pivot(values='Metastatic Sites', index='Timepoint', columns='Drug')

# Preview that Reformatting worked
meta_sites2.head()
```

```python
# Generate the Plot (with Error Bars)
plt.figure(figsize=(13,6))
plt.errorbar(x_axis, meta_sites2["Capomulin"], yerr=meta_sites_se["Capomulin"], linestyle=':', linewidth=1, \
             marker="o", color="red", label="Capomulin")
plt.errorbar(x_axis, meta_sites2["Infubinol"], yerr=meta_sites_se["Infubinol"], linestyle=':', linewidth=1, \
             marker="^", color="blue", label="Infubinol")
plt.errorbar(x_axis, meta_sites2["Ketapril"], yerr=meta_sites_se["Ketapril"], linestyle=':', linewidth=1, \
             marker="s", color="green", label="Ketapril")
plt.errorbar(x_axis, meta_sites2["Placebo"], yerr=meta_sites_se["Placebo"], linestyle=':', linewidth=1, \
             marker="D", color="black", label="Placebo")

plt.grid()
plt.legend(loc="best")
plt.title("Metastatic Spread During Treatment")
plt.ylabel("Met. Sites")
plt.xlabel("Treatment Duration (Days)")
plt.tight_layout()

# Save the Figure
plt.savefig("Images/MetastaticSpread.png")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
surv_rate = study_data.groupby(["Drug", "Timepoint"]).agg({"Mouse ID": "nunique"})\
    .rename(columns={"Mouse ID": "Mouse Count"}).reset_index()

# Preview DataFrame
surv_rate.head(5)
```

```python
# Minor Data Munging to Re-Format the Data Frames
surv_rate2 = study_data.groupby(["Timepoint", "Drug"]).agg({"Mouse ID": "nunique"}).reset_index()\
    .pivot(values='Mouse ID', index='Timepoint', columns='Drug')

# Preview the Data Frame
surv_rate2
```

```python
# generate plot
# each med was tested on 25 mice
surv_rate_calc = (surv_rate2 / 25) * 100
surv_rate_calc

plt.figure(figsize=(13,6))
plt.plot(x_axis, surv_rate_calc["Capomulin"], linestyle=':', linewidth=1, \
             marker="o", color="red", label="Capomulin")
plt.plot(x_axis, surv_rate_calc["Infubinol"], linestyle=':', linewidth=1, \
             marker="^", color="blue", label="Infubinol")
plt.plot(x_axis, surv_rate_calc["Ketapril"], linestyle=':', linewidth=1, \
             marker="s", color="green", label="Ketapril")
plt.plot(x_axis, surv_rate_calc["Placebo"], linestyle=':', linewidth=1, \
             marker="D", color="black", label="Placebo")

plt.grid()
plt.legend(loc="best")
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival rate (%)")
plt.tight_layout()

# Save the Figure
plt.savefig("Images/SurvivalRates.png")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

pct_chg = study_data.groupby(["Timepoint", "Drug"]).agg({"Tumor Volume (mm3)": "mean"}).reset_index()\
    .pivot(values='Tumor Volume (mm3)', index='Drug', columns='Timepoint')\
    .pct_change(axis='columns', periods=9)
pct_chg = pct_chg[45]*100

# Display the data to confirm
pct_chg
```

```python
# Store all Relevant Percent Changes into a Tuple
# Splice the data between passing and failing drugs
pct_chg_plot = pct_chg.loc[pct_chg.index.isin(["Capomulin","Infubinol","Ketapril","Placebo"])]

# Orient widths. Add labels, tick marks, etc. 
#Add Colors
colors = np.where(pct_chg_plot.values <= 0, 'green', 'red')
plt.figure(figsize=(10,7))
bar_lst = plt.bar(pct_chg_plot.index,pct_chg_plot.values, width=0.85, color=colors)
plt.hlines(0,0,3)
plt.ylabel('% Tumor Volume Change')
plt.title('Tumor Change over 45 Day Treatment')

# Use functions to label the percentages of changes
i = 0
for rect in bar_lst:
    if pct_chg_plot[i] <= 0 :
        label_position = -3
    else: 
        label_position = 3
    plt.text(rect.get_x() + rect.get_width()/3, label_position, '%.2f' %  pct_chg_plot[i] + "%")
    i += 1

# Save the Figure 
plt.savefig("Images/Summary.png")
# Show plot
plt.tight_layout()
plt.show()
```
